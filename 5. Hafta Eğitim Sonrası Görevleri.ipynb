
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

# Problem 2

1'den 1000'e kadar olan sayılardan mükemmel sayı olanları ekrana yazdırın. Bunun için bir sayının mükemmel olup olmadığını dönen bir tane fonksiyon yazın.

Bir sayının bölenlerinin toplamı kendine eşitse bu sayı mükemmel bir sayıdır. Örnek olarak 6 mükemmel bir sayıdır (1 + 2 + 3 = 6).

````python
def msayi(sayi):
    bolenler = []
    for i in range(1, sayi):
        if sayi % i == 0:
            bolenler.append(i)
    return sum(bolenler) == sayi

for sayi in range(1, 1001):
    if msayi(sayi):
        print(sayi)

#Bu program, öncelikle mukemmel_sayi adlı bir fonksiyon tanımlar. Bu fonksiyon, aldığı sayının bölünenlerini bulur ve bunların toplamının kendisine eşit olup olmadığını kontrol eder. Eğer toplam kendisine eşitse True, değilse False döndürür.

#Daha sonra program, 1 ile 1000 arasındaki tüm sayıları döngü kullanarak kontrol eder. Her sayı için, mukemmel_sayi fonksiyonunu çağırır ve sayının mükemmel olup olmadığını belirler. Eğer sayı mükemmel ise, print fonksiyonu kullanarak ekrana yazdırır.

#Bu programı çalıştırdığınızda, 1 ile 1000 arasındaki mükemmel sayılar ekrana yazdırılır:
````

# Problem 3

````python
def pisagor(a, b):
    c = (a**2 + b**2) ** 0.5
    return c == int(c)

for a in range(1, 101):
    for b in range(1, 101):
        if pisagor(a, b):
            print(a, b)

#Bir pisagor adlı fonksiyon tanımlayın. Bu fonksiyon, iki tamsayı parametre alacak ve bu sayıların Pisagor üçgeni oluşturup oluşturmadığını belirleyecek. Fonksiyon, True veya False değerini döndürür.

#1 ile 100 arasındaki tüm a ve b değerlerini iki döngü kullanarak kontrol edin.

#Her a ve b çifti için, pisagor fonksiyonunu çağırın ve çiftin Pisagor üçgeni oluşturup oluşturmadığını belirleyin.

#Eğer çift Pisagor üçgeni oluşturuyorsa, ekrana yazdırın.
````

# Problem 4

Bir duvar boyamaya karar verdiniz. Boya kutusunun üzerindeki talimatta, 1 kutu boyanın 5 metrekarelik bir duvarı boyayabileceği yazıyor. Rastgele bir duvar yüksekliği ve genişliği verildiğinde, kaç kutu boya satın almanız gerektiğini hesaplayın.

kutu sayısı = (duvar yüksekliği * duvar genişliği) ÷ kutu başına kaplama.

örneğin Yükseklik = 2, Genişlik = 4, Kaplama = 5

kutu sayısı = (2 * 4) ÷ 5 = 1.6

Ancak bir kutu boyanın 0,6'sını satın alamayacağınız için, sonuç 2 kutuya yuvarlanmalıdır .


````python
yükseklik= int(input(""))
genişlik= int(input(""))

def kutusay(yükseklik,genişlik):
     kutusay= (yükseklik*genişlik) // 5
     print(f"{kutusay} tane kutu boya alınmalı")

kutusay(yükseklik,genişlik)
````

# Problem 5

Bir kelimedeki ünlü ve ünsüz harfleri sayan bir fonksiyon yazınız.

````python
ünlü = "aeıioöuü"
ünlüsayı = 0
ünsüzsayı = 0

def harfsayısı(harfler):
    global ünlüsayı, ünsüzsayı
    for harf in harfler:
        if harf.isalpha():
            if harf in ünlü:
                ünlüsayı += 1
            else:
                ünsüzsayı += 1

harfler = input("Kelime giriniz: ")
harfsayısı(harfler)
print(f"{ünlüsayı} tane ünlü, {ünsüzsayı} tane ünsüz harf var.")
````

# Problem 6

Verilen bir listeden çift sayıları yazdıran bir fonksiyon yazınız.


````python
def cift(sayilar):
    ciftler = []
    for sayi in sayilar:
        if sayi % 2 == 0:
            ciftler.append(sayi)
    return ciftler

sayilar = [1, 2, 3, 4, 5, 6, 7, 8, 9]
cift_sayilar = cift(sayilar)
print(cift_sayilar)
````

# Problem 7

0'dan 10'a kadar olan sayıların toplamını hesaplayan özyinelemeli (Recursive Function) bir fonksiyon oluşturan bir fonksiyon yazınız.

Özyinelemeli (Recursive Function) bir işlev, kendini tekrar tekrar çağıran bir işlevdir.

````python
def toplam(n):
    if n == 0:
        return 0
    else:
        return n + toplam(n-1)

print(toplam(10))

#Burada, toplam adlı fonksiyon, kendisini n sayısına kadar tekrar tekrar çağırarak 0'dan n'ye kadar olan sayıların toplamını hesaplar. İlk olarak, fonksiyonun çağrıldığı sayı sıfır ise, fonksiyon sıfırı döndürür. Aksi takdirde, fonksiyon kendisini n-1 sayısı ile çağırır ve n sayısını bu sonucuyla toplayarak sonucu döndürür.

#Örneğin, toplam(10) çağrısı, 0'dan 10'a kadar olan sayıların toplamını hesaplar ve sonucu döndürür.
````
